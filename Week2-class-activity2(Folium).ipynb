```python
covid_data = pd.read_csv("https://api.covid19india.org/csv/latest/state_wise.csv") 
states_data = pd.read_excel("https://github.com/hrshah211/Corona-Map/raw/main/States.xlsx") 
```

```python
covid_data.head()
```

```python
states_data.head()
```

```python
covid_trunc = covid_data.iloc[1:] 
```

```python
covid_trunc.head()
```

```python
states_data.columns = ['State','Latitude','Longitude'] 
```

```python
states_data.head()
```

```python
merged_data = pd.merge(covid_trunc.sort_values(['State']),states_data.sort_values(['State']))
```

```python
merged_data.head()
```

```python
map = folium.Map(location = (21.146039, 79.085036), zoom_start = 5) 
```

```python
color_set = 'red'
for i in range(len(merged_data)):
    if merged_data.loc[i,'Active'] > 0, tooltip=merged_data.loc[i,'State'], icon=folium.Icon(color=color_set).add_to(map)
```
